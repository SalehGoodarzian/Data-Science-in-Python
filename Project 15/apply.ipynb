### Task 3: add a new column for BMI index
#### BMI for men = 10000 * w / (h ** 2), BMI for women = 12000 * w / (1.1 * h) ** 2

```python
def f1(s, w, h):
    if s == 'M':
        bmi = 10000 * w / (h ** 2)
    else:
        bmi = 12000 * w / (1.1 * h) ** 2
    return round(bmi, 2)
    
df['BMI'] = df.apply(lambda x: f1(x['Sex'], x['Weight'], x['Height']), axis=1)
df
```

### Task 4: add a new column based on BMI index to indicate overall condition
#### BMI<18 is thin, 18<BMI<25 is fit, 25<BMI<30 is overweight, and 30<BMI is fat

```python
def f2(x, a, b, c):
    if x <= a:
        return 'Thin'
    elif x <= b:
        return 'Fit'
    elif x <= c:
        return 'Overweight'
    else:
        return 'Fat'
        
df['Overall condition'] = df['BMI'].apply(lambda x: f2(x, 18, 25, 30))
df
```

### Task 5: men under 40 and women under 35 years of age must attend a training course, find them?

```python
df['Birthdate'] = pd.to_datetime(df['Birthdate'], format='%d/%m/%Y')
df['Age'] = pd.to_timedelta(datetime.now() - df['Birthdate']).dt.days / 365
```

```python
training = df.loc[((df['Sex'] == 'M') & (df['Age'] < 40)) | ((df['Sex'] == 'F') & (df['Age'] < 35))]
training['Employee Name'].tolist()
```

```python
def f1(age, sex, a, b):
    if sex == 'M' and age < a:
        return True
    elif sex == 'F' and age < b:
        return True
    else:
        return False
        
df['Training course'] = df.apply(lambda x: f1(x['Age'], x['Sex'], 40, 35), axis=1)
training = df.loc[df['Training course'] == True]
training['Employee Name'].tolist()
```
